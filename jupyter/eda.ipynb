```python
import pandas as pd
import glob
file_list = glob.glob('./../dataset/*.csv')
file_list
```

```python
clinical = pd.read_csv(file_list[0]).drop(["Unnamed: 0"],axis=1)
print(clinical.shape)
clinical.head()
```

```python
survival = pd.read_csv(file_list[1]).drop(["Unnamed: 0"],axis=1)
print(survival.shape)
survival.head()
```

```python
treatment = pd.read_csv(file_list[2]).drop(["Unnamed: 0"],axis=1)
print(treatment.shape)
treatment.head()
```

```python
genetic_data = pd.read_csv(file_list[3]).drop(["Unnamed: 0"],axis=1)
print(genetic_data.shape)
genetic_data.head()
```

```python
total_data = pd.concat([pd.read_csv(file).drop(["Unnamed: 0"],axis=1) for file in file_list],axis=1)
```

```python
pd.set_option("display.max_colwidth", 1000)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
total_data.head()
```

```python
g_cols = total_data.filter(regex="^G").columns.tolist()
total_data[g_cols] = total_data.filter(regex="^G").astype("category")
```

```python
total_data[["event","Treatment"]] = total_data[["event","Treatment"]].astype("category")
total_data.to_csv("./../processed/total_data.csv",index=False)
```
